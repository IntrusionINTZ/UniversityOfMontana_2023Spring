```python
import pandas as pd
```

```python
data = pd.read_csv('dns_oct-jan.csv')
data
```

```python
# load the whois data set
domains = pd.read_csv('jan26_domain_whois.csv')
domains.head()
```

```python
# merge the two dataframes  
merged = data.merge(domains, left_on='dns_qdomainname', right_on='domain', how='outer')
merged.head()
```

```python
# convert merged data to csv
merged.to_csv('merged_whois.csv')
```
