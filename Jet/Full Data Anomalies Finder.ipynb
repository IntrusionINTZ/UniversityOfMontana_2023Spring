```python
path = 'smallerdata.csv' #this is just the full dataset without some of the columns as shown below
data = pd.read_csv(path)
data
```

```python
pivot_table = data.pivot_table(values = 'shieldid', 
                               index = ['dns_qdomainname', 'day'], 
                               columns = None, 
                               aggfunc = 'nunique') #unstack()
```

```python
pivot_table.reset_index(inplace = True)
```

```python
counts = pivot_table.groupby('dns_qdomainname')['day'].nunique()
```

```python
filtered_df = pivot_table.loc[pivot_table['dns_qdomainname'].isin(counts[counts <= 7].index)]
```

```python
filtered_df
```

```python
filtered_df[filtered_df.shieldid > 5]
```

```python
anomalies = filtered_df[filtered_df.shieldid > 3].set_index('day')
```
