```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('3months_under10_000.csv', index_col='day', parse_dates=True)
data.head()
```

```python
# remove Nans from whois column
data['whois'] = data['whois'].fillna('unknown')
```

```python
# Which rows have Nans?
data[data.isna().any(axis=1)]
```

```python
# fill created, updated, expired nans with 'unknown'
data[['created', 'updated', 'expires']] = data[['created', 'updated', 'expires']].fillna(value=pd.NaT)
```

```python
data.head()
```

```python
# check how many rows have anomoly==1
data[data.anomaly==1]
```

```python
data.shape
```

```python
# copy the anomalys 1000 times to balance the dataframe
rows_to_copy = data[data.anomaly==1]
copies = [rows_to_copy]*1000
data = pd.concat(copies+[data], ignore_index=True)
```

```python
data.shape
```

```python
# tld column
data['tld'] = data.dns_qdomainname.str.split('.').str[1]
```

```python
# column of tlds cross referenced with a list of the top tlds used in malicious activity
mal_tlds = ['live','monster','okinawa','top','boutique','beauty','cn','zone','autos','zw','bd','ke','am','sbs','date','pw','quest','cd','bid','cyou','ml','ws','icu','stream','email','uno','cam','xyz','tokyo','help','casa','rest','win']
data.loc[:,'mal_tld']=0
data.loc[data.tld.isin(mal_tlds),'mal_tld']=1
data.mal_tld.sum()
```

```python
data.head()
```

```python
# convert date columns to date time
date_cols = ['timefirstseen', 'timelastseen', 'created','updated','expires']
for col in date_cols:
    print(col)
    data[col] = pd.to_datetime(data[col],errors='ignore')
```

```python
# create column showing the time between firstseen and lastseen
data['timebetweenseen'] = data['timelastseen']-data['timefirstseen']
```

```python
# put in terms of seconds
data['timebetweenseen'] = data['timebetweenseen']/1000000000
```

```python
data.created = pd.to_datetime(data.created)
```

# Logistic Regression

```python
data.head()
```

```python
# preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# classification model
from sklearn.linear_model import LogisticRegression as logreg

# grid search 
from sklearn.model_selection import GridSearchCV, train_test_split as tts

# metrics
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
X = data[['num_clients', 'num_responses', 'shieldid_count', 
          'rank','shieldid','dns_qdomainname', 'whois', 'tld', 'mal_tld']]
y = data.anomaly
```

```python
X_train, X_test, y_train, y_test = tts(X,y,train_size=0.75)
```

```python
num_features = ['num_clients', 'num_responses', 'shieldid_count', 'rank']
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy=' PUT STRATEGY HERE')),
    ('scaler', MinMaxScaler())
])
    
```

```python
cat_features = ['shieldid','dns_qdomainname', 'whois', 'tld', 'mal_tld']
cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy=' PUT STRATEGY HERE ')),
    ('encoder', OneHotEncoder())
])
```

```python
features_processor = ColumnTransformer(transformers=[
    ('numerical processor', num_pipeline, num_features),
    ('categorical processor', cat_pipeline, cat_features)
], remainder='drop')
```

```python
# classification pipeline
pipe = Pipeline([
    ('feature processor', features_processor),
    ('classifier', logreg)
])
```

```python
# grid search 
param_dic = {'classifier__'}

grid=GridSearchCV(pipe, param_dic, cv=5, scoring=' PUT SCORING HERE ', n_jobs=-1, verbose=1)
```

```python
grid.fit(X_train,y_train)
```

```python
# best parameters
grid.best_params
```

```python
# best pipe
best_pipe = grid.best_estimator
```
