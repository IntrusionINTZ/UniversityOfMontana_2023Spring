```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('set2.csv')
```

```python
data
```

```python
counts=data.pivot_table(values='num_clients',
                       index=data.day,
                       columns='dns_qdomainname',
                       aggfunc=sum,
                       fill_value=0)
counts
```

```python
countsT = counts.T
```

```python
#number of consecutive zeros
v = countsT.stack()
m = v.eq(0)

consec_0=(m.ne(m.shift())
  .cumsum()
  .where(m)
  .dropna()
  .groupby(level=0)
  .value_counts(sort=False)
  .groupby(level=0)
  .apply(list))
```

```python
consec_0
```

```python
consec_0['0-02.net']
```

```python
from scipy.signal import find_peaks
```

```python
dom=counts['000webhostapp.com']
peaks,locs = find_peaks(dom,
                     height=1,
                     #threshold=1,
                     #distance=1,
                     prominence=1,
                     #width=1,
                     #wlen=2,
                     #rel_height=1,
                     #plateau_size=1
                    )
plt.plot(counts['000webhostapp.com'])
plt.plot(peaks,dom[peaks],'x')
```

```python
np.diff(peaks)
```

```python
goog=counts['google.com']
peaks,_ = find_peaks(goog,
                     height=50000,
                     #threshold=1,
                     distance=1,
                     
                     width=1,
                     #wlen=2,
                     #rel_height=1,
                     #plateau_size=1
                    )
plt.plot(goog)
plt.plot(peaks,goog[peaks],'x')
```

```python
np.diff(peaks)
```

```python
doms = counts.columns
doms
```

```python
delta_peaks=[]
for i in doms:
    dom=counts[i]
    peaks,_ = find_peaks(dom,
                     #height=1,
                     #threshold=1,
                     #distance=1,
                     prominence=1,
                     #width=1,
                     #wlen=2,
                     #rel_height=1,
                     #plateau_size=1
                    )
    delta_peaks.append(np.diff(peaks))
delta_peaks 
```
